```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.array(range(1,7))
good_dice = np.array([1/6]*6)
bad_dice = np.array([1/10]*5 + [1/2])
p1 = np.float128(0.04)
p2 = np.float128(0.05)
```

```python
def roll_dice(n, good=True):
    res = []
    colors = []
    while n > 0:
        dice = good_dice if good else bad_dice
        res.append(np.random.choice(results, size=1, p=dice)[0])
        colors.append(0 if good else 1)
        n -= 1
#         print(good)
        if good and np.random.rand() <= p1:
            good = False
            continue
        if (not good) and np.random.rand() <= p2:
            good = True
    return np.array(res), np.array(colors)
```

```python
# function by Michał Zobniów
def plot_seq(roll_seq, color):
    if len(roll_seq) > 10000:
        l = 5000
    else:
        l = len(roll_seq)
    colors = ['g', 'r', 'black']
    plt.figure(figsize=(40,5))
    timeTM = np.linspace(0, l/2, l)
    for i, x in enumerate(timeTM): 
        plt.plot([x, x], [0, roll_seq[i]], c=colors[color[i]], )
    plt.xlim(0, 500)
    plt.ylim(0,6)
    plt.show()
```

```python
def solve_with_h(results):
    res = []
    n = len(results)
    step = 9
    for i in range(0,n,step):
        expected = sum([1 if e == 6 else 0 for e in results[i:i+step]])
        if expected >= step/3:
            res += [1]*len(results[i:i+step])
        else:
            res += [0]*len(results[i:i+step])
    return np.array(res)
```

```python
def score(target, label):
    assert len(target) == len(label), str(len(target)) + ' ' + str(len(label))
    s = np.sum(target == label)
    return s/len(target)
```

```python
def a(i, j):
    if j == 0:
        if i == 1:
            return 1
        else:
            return 0
    s = [p1, p2]
    if i == 1:
        if j == 1:
            return 1-s[0]
        else:
            return s[0]
    else:
        if j == 2:
            return 1-s[1]
        else:
            return s[1]

        
def b(i,j,k):
    dice = good_dice if j == 1 else bad_dice
    return dice[k-1]

def alpha(o):
    res = [[a(1,0)], [a(2,0)]]
    for t in range(1, len(o)+1):
        res[0].append(sum(res[i-1][-1]*a(i,1)*b(i,1,o[t-1])*5 for i in range(1,3)))
        res[1].append(sum(res[i-1][-1]*a(i,2)*b(i,2,o[t-1])*5 for i in range(1,3)))
    return np.array(res)
    
    
def beta(o):
    res = [[1],[1]]
    for t in range(len(o), 0, -1):
        res[0].insert(0, sum(a(1,j) * b(1,j,o[t-1]) * res[j-1][0] *5 for j in range(1,3)))
        res[1].insert(0, sum(a(2,j) * b(2,j,o[t-1]) * res[j-1][0] *5 for j in range(1,3)))
    return np.array(res)

def solve_with_viterbi(o):
    gamma = alpha(o) * beta(o) 
    return np.array([0 if gamma[0][i] > gamma[1][i] else 1 for i in range(len(gamma[0]))][1:])
```

```python
N=10000
res, color = roll_dice(N)
plot_seq(res, color)
```

```python
heur = solve_with_h(res)
plot_seq(res, heur)
```

```python
score(color, heur)
```

```python
vit = solve_with_viterbi(res)
plot_seq(res, vit)
```

```python
score(color, vit)
```
